```python
#2.10
n_horas = int(input(f'\nQuantas horas você trabalhou esse mês? '))
v_hora = float(input(f'\nQual o valor contratual da sua hora de trabalho? '))

def salario_bruto ():
    return n_horas * v_hora
    
def irpf ():
    if salario_bruto() < 5000:
        irpf = 0
    elif salario_bruto() > 7350:
        irpf = salario_bruto() * 0.275 - 908.73 
    else:
        irpf = salario_bruto() * 0.275 - 908.73 - (978.62 - (0.133145 * salario_bruto()))
    return irpf

def salario_liq ():
    return salario_bruto() - irpf()

print(f"\nConsiderando essas informações, seu salário bruto será {salario_bruto()}, seu IRPF será de {irpf()} e seu salário líquido será {salario_liq()}")
```

```python
#3.1
def polinomio(x):
    polinomio = 0.0
    n = int(input('Informe o grau do polinômio que deseja calcular: '))
    coeficientes = []
    i = 0
    while i < n:
        a = float(input(f'\nInforme o coeficiente de {i+1}º: '))
        coeficientes.append(a)
        i+=1
    
    i = 0 #reinicia a contagem do 0
    while i < len(coeficientes):
        polinomio += coeficientes[i] * (x**(i+1))
        i+=1
    print(polinomio)
polinomio(2)
```

```python
#3.2
a = float(input('Informe um número: '))
if a < 0:
    print("Esse número é inválido para o cálculo de raiz")
else:
    x = a / 2
    # Tolerância (critério de parada)
    tolerancia = 0.0001
    while True:
        x_novo = 0.5 * (x + a / x)
        if abs(x_novo - x) < tolerancia:  # diferença pequena o suficiente → para
            break
        x = x_novo
print(f"Raiz quadrada de {a} ≈ {x_novo:.4f}")    
```
