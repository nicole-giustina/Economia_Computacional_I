```python
"""
Programa: Adulto_menor
Descrição: este programa pergunta a idade de uma pessoa e, dependendo da idade, apresenta a mensagem 'Oi! Você é um adulto.' ou 'Oi! Você é menor de idade'.
Nome: Nicole Schultz
Data: 24/03/2026
Versão 0.0.1
"""
```
