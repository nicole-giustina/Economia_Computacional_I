# Processamento de arquivos

```python
# Leitura de arquivo

arquivo001 = open('arquivo001.txt', 'r') 

# Preciso colocar em tese o caminho completo até o arquivo. 
# Aqui estou usando o nome do arquivo direto pois estou na mesma pasta; ainda, devo escolher o modo por meio do qual vai ser aberto, posso consultar isso na biblioteca python dos códigos. 
# Os mais usados são r, w, a. O r inclusive é o padrão, se fechar o parênteses logo depois do nome do arquivo, será essa a função utilizada.

arquivo001.close()
# Sempre que eu abrir um arquivo, no fim do código ou depois de usar, deve escrever também o código para fechar para não tem problemas. 
# O ideal é escrever o open e o close juntos e escrever o que quiser no meio.
```

```python
arquivo002 = open('arquivo002.txt', 'w') #criar um arquivo com esse nome na função write

for linha in range(1, 101): 
    arquivo002.write('%d\n' % linha) #agora para cada linha de 1 até 100 (a última não pega), você vai substituir o %d por linha; \n é o return por meio de código, ou seja, próxima linha; 
    #e ele vai iterando até o último, que é o 100 (intervalo entre 1 e 101 pega o primeiro e exclui o último)

arquivo002.close()
```

```python
arquivo003 = open('arquivo003.txt', 'w')

arquivo003.write('casa') #o método write escreve qualquer string no objeto indicado, só escreve strings na verdade, não funções. Se colocar 2+2 ele grava exatamente isso, e não 4. 
# É análogo ao print, mas é um método de um arquivo, e não uma função.

arquivo003.close()
```

## Um outro jeito de usar a função open, como gerenciador de contexto with
<!-- um gerenciador de contexto gerencia memória, ao invés de eu precisar fazer isso manualmente abrindo e fechando sempre --> 

```python
with open('arquivo004.txt', 'w') as arquivo004obj:
    #agora tudo que estiver endentado é um contexto do with, de modo que não preciso me preocupar em fechar, ele cuida de tudo que estiver aqui dentro. Posso até colocar um with dentro do outro.
    arquivo004obj.write('casa')
```

## Criando um csv

```python
with open('arquivo005.csv', 'w') as arquivo005:
    i=0
    while i<11:
        arquivo005.write(f'{i}, {2*i}\n') #aqui tenho um jeito mais moderno de escrever a mesma coisa que no arquivo001. 
        # como ele sabe escrever str, na primeira linha ele vai escrever 1,2 e quebra a linha, depois 2,4 e assim por diante. A vírgula separa coluna em csv. No formato brasileiro separa por ;
        i+=1
```
