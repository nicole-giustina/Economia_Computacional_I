```python
# Classe Restrição orçamentária
"""
Esta classe recebe a renda m, o preço do bem 1 p1 e do bem 2 p2, e avalia se uma cesta genérica (x1, x2) é financeiramente viável.
"""

class Restricao_orcamentaria:
    def __init__(self, m, p1, p2):
        self.renda = m #Renda
        self.p1 = p1 #Preço bem 1
        self.p2 = p2 #Preço bem 2

    def viabilidade(self, x1, x2):
        rest_orc = (self.p1 * x1) + (self.p2 * x2)
        return self.renda >= rest_orc

cesta = Restricao_orcamentaria(100, 20, 10)
cesta.viabilidade(3, 5)
```
