```python
def reader():
    i = 0
    dados = [0,0]
    while i < 2:
        dados[i] = float(input(f'Digite a parcela {i+1}'))
        i+=1
    return dados    
```

```python
def impressora(dados: list, resultado: float) -> None:
    """ Esta função imprime na tela """
    print(f'A soma dos números {dados[0]} e {dados[1]} é igual a {resultado}.')
```

```python
def soma(parcela1: float, parcela2: float) -> float:
    """ Esta função soma dois números reais """
    resultado = parcela1 + parcela2
    return resultado
```

```python
def main():
    dados = reader()
    resultado = soma(dados[0], dados[1])
    impressora(dados, resultado)
```

```python
main()
```
